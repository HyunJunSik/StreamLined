```python
import matplotlib.pyplot as plt

training_time = [
    "CLKD's batch time : 152.301ms",
    "MLKD's batch time : 155.931ms",
    "SDD's batch time : 159.343ms",
    "Fitnet's batch time : 152.329m",
    "RKD's batch time : 171.689ms",
    "OFD's batch time : 184.718ms",
    "ReviewKD's batch time : 181.030ms",
    "CRD's batch time : 187.165ms",
    "Ours's batch time : 150.659ms",
]

names = []
times = []
for i in range(len(training_time)):
    time = float(training_time[i][-3:-10:-1][::-1])
    name = training_time[i].split(" ")[0][-3::-1][::-1]
    times.append(time)
    names.append(name)

colors = [
    '#d0e7ff',  
    '#7faadf',
    '#5b8ccc',
    '#a4c8f0',
    '#3a6eb9',
    '#2b5aa4',
    '#1f478f',
    '#123374'   
]
plt.figure(figsize=(4,2))
bars = plt.bar(names, times, color=colors)

plt.ylabel("Training Time (ms)", fontsize=8, fontweight='bold')
plt.xticks(rotation=45, fontsize=8, fontweight='bold')
plt.yticks(fontsize=8, fontweight='bold')
plt.ylim(145, max(times) + 10)

plt.tight_layout()
plt.show()
```
